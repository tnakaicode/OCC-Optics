```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_logistc (px):
    sigmo=2
    maxim=np.sqrt(10) - 1
    ratio=10
    py = maxim / (1 + np.exp( -ratio*( px - sigmo)))

    plt.figure()
    plt.plot (px, py)
```

```python
px = np.linspace (-1, 1,100)
plot_logistc(px)
plt.show()
```
