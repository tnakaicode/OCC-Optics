```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func_logistc (px=0, cff=2.0, rng=1.0, sig=0):
    py = cff / (1 + np.exp( -( px - sig)/rng))
    return py
```

```python
px = np.linspace (-1, 1,100)*100

plt.figure()
plt.grid()
plt.plot(func_logistc(px, sig=0.0))
plt.plot(func_logistc(px, sig=0.0, rng=10.0))
plt.plot(func_logistc(px, sig=0.0, rng=10.0, cff=3.0))
plt.show()
```
